### Imports

Fist we import the packages we will be using and then we import our csv into a SFrame

```python
import turicreate as tc
import math
import operator
```

```python
activity_data = tc.SFrame('activity.csv')
activity_health = activity_data.apply(lambda row: (row['Healthy Physically'] + row['Healthy Intellectually'] + row['Healthy Mentally'])/3)
activity_data = activity_data.add_column(activity_health, 'Activity Health')
activity_data
```

### Modelling

We define the cosine similarity function that we will be using to model the similarity between different activities

```python
def cosine(u, v):
    '''
    Given two vector inputs u, v compute the cosine similarity between the two vectors
    '''
    xx, xy, yy = 0, 0, 0
    for z in u:
        if type(u[z]) is str: # skip first two elements which are strings
            continue
        xx += u[z] * u[z]
        xy += u[z] * v[z]
        yy += v[z] * v[z]
    return xy/math.sqrt(xx*yy)
```

### Computation

For each activity we compute its cosine similarity to all other activities

```python
similarities = {} # dictionary for computed similarity values
for row in activity_data:
    similarities[row['Activity']] = {}
    for i in range(len(activity_data)):
        if (row['Activity'] == activity_data[i]['Activity']):
            continue
        similarities[row['Activity']][activity_data[i]['Activity']] = cosine(row, activity_data[i])
```

### User Input

Here a user inputs the activities that he or she enjoys practicing and we compute a user health score based on those activities.

The user also inputs the amount of free periods they have throughout the day.

```python
user_activities = ['Jog', 'Walk', 'Yoga'] # Edit this to input your own activities
user_timings = { # Edit the numbers to indicate how many free periods of a certain duration you have throughout a day
    "<15 min": 1,
    "15 to 30": 3,
    "30 to 60": 2,
    "hour+": 1
}
user_health_score = 0
for i in range(len(user_activities)):
    user_health_score += activity_data[activity_data['Activity'] == user_activities[i]]['Activity Health'][0]
user_health_score = user_health_score / (3 * len(user_activities))
```

Using the user's health score and each activity's health score we slightly weight the similarity score previously computed.

(If a person has a health score of 1 then no ajudstment is made, if their health score is 0 then it is adjusted by 0.25 * activity health score)

```python
for activity in similarities:
    for similar in similarities[activity]:
        similarities[activity][similar] = (0.25 - 0.25 * user_health_score) * activity_data[activity_data['Activity'] == similar]['Activity Health'][0] + (0.75 + 0.25 * user_health_score) * similarities[activity][similar]  
```

```python
# uncomment to see the dictionary of similarities
# print(similarities)
```

To make it easier to retrieve the most similar activities we create a new dictionary of similarities which is orders the keys by their values in descending order.

Thus for any given activity in the ordered_similarities dictionary, the first key-value pair of the embeded dictionary denotes the activity it is most similar to.

```python
ordered_similarities = {}
for activity in similarities:
    ordered_similarities[activity] = sorted(similarities[activity].items(), key=operator.itemgetter(1), reverse=True)
```

For each activity we then retrieve the top ten most similar activities

```python
top_ten_dict = {'Activity': [], 'Related Activity': [], 'Similarity Score': []}
for activity in ordered_similarities:
    for i in range(10):
        top_ten_dict['Activity'].append(activity)
        top_ten_dict['Related Activity'].append(ordered_similarities[activity][i][0])
        top_ten_dict['Similarity Score'].append(str(ordered_similarities[activity][i][1]))
top_ten = tc.SFrame(top_ten_dict)
top_ten
```

### Making suggestions to a user

Having established the most similar activities (potentially slightly adjusted for health) we establish the activities most similar to those the user practives and we make suggestions to him or her.

```python
similar_activities = []
similar_activities = set(similar_activities);
for user_activity in user_activities:
    for row in top_ten[top_ten['Activity'] == user_activity]:
        if row['Related Activity'] not in user_activities and row['Similarity Score']:
            similar_activities.add(row['Related Activity'])
print(similar_activities)
```

Suggestions come in the form of a dictionary which presents a list of recommended activities for each specific time period.



```python
activity_per_period = {
    "<15 min": [],
    "15 to 30": [],
    "30 to 60": [],
    "hour+": []
}

suggestions_made = set([])
for period in user_timings:
    for row in activity_data[activity_data[period] > 0].sample(0.5):
        if row['Activity'] in similar_activities and len(activity_per_period[period]) < user_timings[period] + 1 and row['Activity']:
            activity_per_period[period].append(row['Activity'])
            suggestions_made.add(row['Activity'])

activity_per_period
```
